```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
```

```python
#Вивести перші рядки таблиці за допомогою методу head
df.head(5)
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
lines, columns = df.shape
print(f'Number of lines: {lines}')
print(f'Number of columns: {columns}')
```

```python
#Замініть у таблиці значення "—" на значення NaN
df = df.replace({'—': np.NaN})
df.head(5)
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
nan_columns = df.columns[df.isnull().any()].tolist()
df[nan_columns] = df[nan_columns].astype(float)

df.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() / len(df)  
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.drop([27], axis=0)
df.tail(5)
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df = df.fillna(df.mean(numeric_only=True), axis=0)
df.head(5)
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
avg = df['2019'].mean()
df.loc[df['2019'] > avg]
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
max_value_region = df[df['2014'] == df['2014'].max()]['Регіон']
max_value_region
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
sns.set_style('darkgrid')

plt.bar(df['Регіон'], df['2019'])
plt.xlabel("Область")
plt.ylabel("Значення")
plt.title("Народжуваність по регіонах у 2019 році", fontsize=15)

plt.tick_params(labelrotation = 90)

plt.show()
```

```python
labels = ['Вінницька', 'Волинська', 'Дніпропетровська', 'Донецька']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Народжуваність в регіонах України у 2019 році')

plt.show()
```

```python
categories = ['2000', '2012', '2014', '2020']
values = [7.3, 12.7, 15, 9]
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='purple', alpha=0.7)
plt.title('Коефіцієнт народжуваності в Кримy')
plt.xlabel('Роки')
plt.ylabel('Значення')
plt.grid(axis='y')
```

```python
x3 = np.linspace(0, 4 * np.pi, 50)
y3 = np.sin(x3)
plt.figure(figsize=(10, 4))
plt.plot(x3, y3, label='y = sin(x)', color='green', linestyle='-.')
plt.title('Коефіцієнт народжуваності в Житомирській обл.')
plt.xlabel('роки')
plt.ylabel('народжуваность')
plt.grid(True)
plt.legend()
```
