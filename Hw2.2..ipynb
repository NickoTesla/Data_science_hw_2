```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

cnt = df.isnull().sum(axis=0)*100/8704
cnt.round(3)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця Язык.программирования

df_tmp = df.isna().any()
col_not_del = ['Язык.программирования']
df_tmp[col_not_del] = False
df.drop(df_tmp.index[df_tmp], axis=1, inplace=True)    
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець 
# "Мова.програмування"

cnt = df.isnull().sum(axis=0)*100/8704
cnt.round(3)
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

df = df.dropna()
df.head()
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
df.shape
```

```python
# Створіть нову таблицю python_data з рядками, де "Мова програмування" - Python
python_data = df[df["Язык.программирования"] == "Python"]
```

```python
# Визначте розмір таблиці python_data
print("Розмір таблиці python_data:", python_data.shape)
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", 
# виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у 
# стовпчику "Зарплата.в.місяць"

df_new = df
df_new = df.groupby(['Должность'])['Зарплата.в.месяц'].agg([np.min, np.max])
df_new
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик "avg"


def fill_avg_salary(salary):
    return salary.mean()

df_new['avg'] = df_new.apply(fill_avg_salary, axis=1)
df_new
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.

df_new['avg'].describe()
```

```python
# Збережіть отриману таблицю в CSV файл

df_new.to_csv('2017_jun_final_new.csv', encoding='utf8')
```
