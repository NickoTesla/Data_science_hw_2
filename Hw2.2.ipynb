```python
# Порахуйте частку пропусків в кожному стовпці
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Частка пропусків в кожному стовпці:")
print(missing_percentage)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
df = df.dropna(subset=["Мова програмування"])
```

```python
# Порахуйте знову частку пропусків в кожному стовпці
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Частка пропусків після видалення:", missing_percentage)
```

```python
# Видаліть усі рядки з пропусками
df = df.dropna()

```

```python
# Визначте новий розмір таблиці
print("Розмір таблиці після видалення рядків:", df.shape)
```

```python
# Створіть нову таблицю python_data з рядками, де "Мова програмування" - Python
python_data = df[df["Язык.программирования"] == "Python"]
```

```python
# Визначте розмір таблиці python_data
print("Розмір таблиці python_data:", python_data.shape)
```

```python
# Виконайте групування за стовпчиком "Посада" і знайдіть мінімальну та максимальну зарплату в місяць
aggregated_data = df.groupby("Должность")["Зарплата.в.месяц"].agg(["min", "max"])
```

```python

# Створіть функцію fill_avg_salary, яка повертає середнє значення заробітної плати на місяць
def fill_avg_salary(group):
    return group.mean()
```

```python
# Використайте її для створення нового стовпця "avg"
aggregated_data["avg"] = df.groupby("Должность")["Зарплата.в.месяц"].apply(fill_avg_salary)
```

```python
# Виведіть описову статистику для нового стовпця
print(aggregated_data.describe())
```

```python
# Збережіть отриману таблицю в CSV файл
aggregated_data.to_csv("aggregated_data.csv")
```
